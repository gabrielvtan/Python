# Matplotlib Basics -1 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# this command allows you to see the graph inline without using plt.show()
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# Functional method of showing plot (not better than OOP)
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiple plots on same canvas
# first is number of rows, then number of columns, and the plot you want to show
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method for visualization
# create figure objects then make methods on it
fig = plt.figure()

# add_axes() takes 4 arguments, left, bottom, width, height in list form
ax = fig.add_axes([0.1,0.1,0.8,0.8])

# now we plot on the axes
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

```python
# Two sets of figures on one canvas
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
# As you can see, this creates a second subplot within the existing figure
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)
ax2.plot(y,x)

ax1.set_title('Large Plot')
ax2.set_title('Small Plot')
```

# Matplotlib Basics - 2

```python
# nice way to create multiple figures and axes
fig,axes = plt.subplots(nrows=3, ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#Since axes is simply a list of matplotlib objects, we can iterate through it

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

# Figure size and DPI

```python
# DPI is dot/pixels per inch
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows =2, ncols=1,figsize=(8,2))

ax[0].plot(x,y)
ax[1].plot(y,x)
plt.tight_layout()
```

```python
# Save a figure
# bigger dpi increases the size of the figure
fig.savefig('my_picture.png', dpi=200)
```

```python
# Legends
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

# See location code documentation for proper legend spacing
# 0 = best
ax.legend(loc=10)
```

# Matplotlib Basics - 3

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--', lw=1, marker = 'o', markersize=20) # green dashed line
```

```python
fig, ax = plt.subplots()

ax.plot(x,x+1, color='blue', alpha=0.5)
ax.plot(x,x+2, color='#8B008B')
ax.plot(x, x+3, color="#FF8C00") 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+7, color="green", lw=3, ls=':')
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60]) # here we set a lower and upper bound for the y axis
axes[2].set_xlim([2, 5]) # here we set a lower and upper bound for the x axis
axes[2].set_title("custom axes range");
```
