## Using Statsmodels to get the trend
The Hodrick-Prescott filter separates a time-series  y_t  into a trend  τ_t and a cyclical component  ζt

$y_t = \tau_t + \zeta_t$

The components are determined by minimizing the following quadratic loss function

$\min_{\\{ \tau_{t}\\} }\sum_{t}^{T}\zeta_{t}^{2}+\lambda\sum_{t=1}^{T}\left[\left(\tau_{t}-\tau_{t-1}\right)-\left(\tau_{t-1}-\tau_{t-2}\right)\right]^{2}$

```python
# the hpfilter specific returns a tuple representing the estimated seasonality(cyclical nature) and trend of the data 
gdp_cycle, gdp_trend = sm.tsa.filters.hpfilter(df['realgdp'])
```

```python
df['trend'] = gdp_trend
df['cycle'] = gdp_cycle
```

```python
df[['realgdp', 'trend']]['2000-03-31':].plot()
```

# EWMA Theory

```python
# EWMA - Exponentially Weighted Moving Averages
# This model improves upon SMAs shortcomings by allowing u to reduce the lage effect from SMA and it will put more
# weight on values that occurred more recently
```

```python
airline = pd.read_csv("airline_passengers.csv", index_col="Month")
# Since the month data is a string - we have to make some data manipulations
airline.index
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
# Here is an example of a SMA
airline['6-month-SMA']=airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12-month-SMA']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.plot(figsize=(10,8));
```

```python
# Here is an example of an EWMA, where we put more weight on the more recent data points relative the earlier ones
# There are multiple parameters to consider - but we will focus on 'span' which refers to 'N-day EW moving average'
airline['EWMA-12']=airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers', 'EWMA-12']].plot(figsize=(10,8));
```
